The Silhouette Score is a metric used to evaluate the quality of a
clustering solution. It measures how similar each data point is to 
its own cluster compared to other clusters. The Silhouette Score can 
help determine how well-defined and distinct the clusters are.

# Step-by-Step Calculation of the Silhouette Score

Mean Intra-Cluster Distance (a)

For each data point 
i, calculate the average distance between 
i and all other points in the same cluster. This is known as the intra-cluster distance 
a(i).

Mean Nearest-Cluster Distance (b)

For each data point 
i, calculate the average distance between 
i and all points in the nearest cluster to which 
i does not belong. This is known as the nearest-cluster distance 
b(i).

# Interpretation

+1: Indicates that the data point is well clustered and far away from the neighboring clusters.

0: Indicates that the data point is on or very close to the decision boundary between two neighboring clusters.

-1: Indicates that the data point might have been assigned to the wrong cluster, as it is closer to a neighboring cluster than to the cluster it is assigned to.

```python
import numpy as np
from sklearn.metrics import pairwise_distances

# Step 1: Calculate pairwise distances between all points
D = pairwise_distances(data)

# Step 2: Initialize silhouette scores
silhouette_vals = []

# Step 3: Loop through each point to compute a(i) and b(i)
for i in range(len(data)):
    # Cluster label of the current point
    current_cluster = r[i]

    # All indices in the current cluster (excluding point i)
    in_cluster = np.where(r == current_cluster)[0]
    in_cluster = in_cluster[in_cluster != i]

    # Compute a(i): mean distance to all other points in same cluster
    if len(in_cluster) > 0:
        a_i = np.mean(D[i][in_cluster])
    else:
        a_i = 0  # Cluster has only one point

    # Compute b(i): minimum mean distance to points in other clusters
    b_i = np.inf
    for cluster_id in np.unique(r):
        if cluster_id == current_cluster:
            continue
        other_cluster = np.where(r == cluster_id)[0]
        b_i = min(b_i, np.mean(D[i][other_cluster]))

    # Compute silhouette score for point i
    s_i = (b_i - a_i) / max(a_i, b_i) if max(a_i, b_i) > 0 else 0
    silhouette_vals.append(s_i)
    
    

# Step 4: Compute the average silhouette score
manual_silhouette_score = np.mean(silhouette_vals)

print("Manual Silhouette Score:", manual_silhouette_score)
```

The Davies-Bouldin Score (DB score) is a metric for evaluating the quality of
a clustering solution. It is based on the average similarity ratio of each
cluster with the cluster that is most similar to it. Lower Davies-Bouldin 
scores indicate better clustering performance

```python
# Calculate Davies-Bouldin Score
kmeans_db_score = davies_bouldin_score(data,r)
kmeans_db_score
```
